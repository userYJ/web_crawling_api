```python
import requests
import pandas as pd
```

### headers : daum exchange rate : 환율 데이터 수집 (국가별 환율 데이터)

```python
# 1. 웹 서비스 분석 : URL
```

```python
url = f"https://finance.daum.net/api/exchanges/summaries"
```

```python
# 2. request(url) > response(data) : data(json(str))
```

```python
headers = {
    "referer": "https://finance.daum.net/exchanges",
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36 Edg/98.0.1108.56"
} # user- agent가 파이썬으로 찍히므로 이걸 바꿔줌

response = requests.get(url, headers = headers)
response
```

```python
# 3. data(json(str)) > dict, list > df
```

```python
datas = response.json()["data"]
df = pd.DataFrame(datas)[["symbolCode","date","currencyCode","basePrice"]]
df
```
